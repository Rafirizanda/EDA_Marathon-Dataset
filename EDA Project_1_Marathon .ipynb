```python
#Step 1 : Show 46K or 46Mi
#46 Km

df[df['Event distance/length'] == '46km']
```

```python
#Combine 46k/46mi with isin

df[df['Event distance/length'].isin(['46km', '46mi'])]
```

```python
#Combine all the filters together
df[(df['Event distance/length'].isin(['46km', '46mi'])) & (df['Year of event'] == 2018) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'FRA')]
```

```python
df2 = df[(df['Event distance/length'].isin(['46km', '46mi'])) & (df['Year of event'] == 2018) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'FRA')]
```

```python
df2.head(10)
```

```python
df2.shape
```

```python
#remove (FRA) from event name
df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
#Clean up athlete age
df2['athlete_age'] = 2018 - df2['Athlete year of birth']
```

```python
#remove h from athlete performance
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(5)
```

```python
#drop columns : Athlete club, Athlete country, Athlete year of birth, Athlete age category
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
df2.head(5)
```

```python
#Clean up null values (find null)
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
#Check duplicate values
df2[df2.duplicated() == True]
```

```python
#Reset index
df2.reset_index(drop = True)
```

```python
#Fix types
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2.head()
```

```python
#Rename columns
#Year of event                  int64
#Event dates                   object
#Event name                    object
#Event distance/length         object
#Event number of finishers      int64
#Athlete performance           object
#Athlete gender                object
#Athlete average speed        float64
#Athlete ID                     int64
#athlete_age                    int32
```

```python
df2 = df2.rename(columns = {'Year of event': 'year', 
                        'Event dates': 'race_day',
                        'Event name': 'race_name',
                        'Event distance/length': 'race_length',
                        'Event number of finishers': 'finish_number',
                        'Athlete performance': 'athlete_performance',
                        'Athlete gender': 'athlete_gender',
                        'Athlete average speed': 'athlete_average_speed', 
                        'Athlete ID': 'athlete_id',                  
                        })
```

```python
df2.head()
```

```python
#reorder columns
df3 = df2[['race_day', 'race_name', 'race_length', 'finish_number', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed']]
```

```python
df3.head()
```

```python
#The questions i want to find out from the data
```

```python
#What age groups are the best in the 46km race (20+ race minimum)
df3.query('race_length == "46km"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count>19').head(20)
```

```python
#What age groups are the worst in the 46km race (20+ race minimum)
df3.query('race_length == "46km"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('count>19').head(20)
```

```python
#3. Is there a difference in average speed (Athlete average speed) between athlete genders in each year of the event (Year of event)?
df3.query('race_length == "46km"').groupby('athlete_gender')['athlete_average_speed'].agg(['mean']).sort_values('mean', ascending = False)
```

```python
#Which race (race_name) has the most finishers (Event number of finishers)
df3.groupby('race_name')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
